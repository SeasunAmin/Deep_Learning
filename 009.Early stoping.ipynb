## Early stoping for know the excet numbers of epoch we shoule use

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#making a dataset

from sklearn.datasets import make_circles

(x,y) = make_circles(random_state=1,noise=0.1)
```

```python
sns.scatterplot(x[:,0],x[:,1],hue=y)
```

```python
#spliting the dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=1)
```

```python
#creating the model

model = Sequential()

model.add(Dense(256,activation="relu",input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy", metrics=['accuracy'],optimizer='Adam')
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),verbose=0,epochs=3000)
```

```python
plt.plot(history.history['loss'],label = "Train")
plt.plot(history.history['val_loss'],label = "Test")

plt.legend()
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x_test,y_test.ravel(),clf=model,legend=2)
plt.show()
```

# Applying early stoping

```python
model = Sequential()

model.add(Dense(256,activation="relu",input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy", metrics=['accuracy'],optimizer='Adam')
```

```python
#early stoping


from tensorflow.keras.callbacks import EarlyStopping

callback = EarlyStopping(monitor="val_loss",
    min_delta=0.00001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False)
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3000, callbacks=callback)
```

```python
plt.plot(history.history['loss'],label = "Train")
plt.plot(history.history['val_loss'],label = "Test")

plt.legend()
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x_test,y_test.ravel(),clf=model,legend=2)
plt.show()
```
